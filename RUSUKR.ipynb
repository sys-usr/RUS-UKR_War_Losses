```python
import pandas as pd
import matplotlib.pyplot as plt

rus_l = pd.read_csv('losses_russia.csv')

display(rus_l.describe())

display(rus_l.info())

display(rus_l.head(10))
```

```python
ukr_l = pd.read_csv('losses_ukraine.csv')

display(ukr_l.describe())

display(ukr_l.info())

display(ukr_l.head(10))
```

```python
columns_to_analyze = ['equipment','model','sub_model','manufacturer','losses_total','abandoned','abandoned and destroyed','captured','captured and destroyed','damaged','damaged and abandoned','damaged and captured','damaged beyond economical repair','destroyed','scuttled to prevent capture by Russia','sunk','sunk but raised by Russia']

# Compute the correlation matrix
corr_matrix = ukr_l[columns_to_analyze].corr()

# Iterate over the correlation matrix and print the correlated categories
for col1 in corr_matrix.columns:
    for col2 in corr_matrix.columns:
        if col1 != col2:
            correlation = corr_matrix.loc[col1, col2]
            if correlation >= 0.5 or correlation <= -0.5:
                print(f"Correlation between {col1} and {col2}: {correlation}")
```

```python
# Get the filenames without the extension
filename1 = 'losses_russia'
filename2 = 'losses_ukraine'

# Read the CSV files into separate DataFrames
df1 = pd.read_csv(filename1 + '.csv')
df2 = pd.read_csv(filename2 + '.csv')

# Add the filename prefix to each column in df1
df1_renamed = df1.add_prefix(filename1)

# Add the filename prefix to each column in df2
df2_renamed = df2.add_prefix(filename2)

# Concatenate the renamed DataFrames along the columns axis (axis=1)
concatenated_df = pd.concat([df1_renamed, df2_renamed], axis=1)

# Save the concatenated DataFrame to a new CSV file
concatenated_df.to_csv('concatenated.csv', index=False)
```

```python
RUS_UKR = pd.read_csv('concatenated.csv')


RUS_UKR_Columns = [
    'losses_russiaequipment', 'losses_russiamodel', 'losses_russiasub_model', 'losses_russiamanufacturer',
    'losses_russialosses_total', 'losses_russiaabandoned', 'losses_russiaabandoned and destroyed',
    'losses_russiacaptured', 'losses_russiacaptured and destroyed', 'losses_russiacaptured and stripped',
    'losses_russiadamaged', 'losses_russiadamaged and abandoned', 'losses_russiadamaged and captured',
    'losses_russiadamaged beyond economical repair', 'losses_russiadestroyed',
    'losses_russiadestroyed in a non-combat related incident', 'losses_russiasunk',
    'losses_ukraineequipment', 'losses_ukrainemodel', 'losses_ukrainesub_model', 'losses_ukrainemanufacturer',
    'losses_ukrainelosses_total', 'losses_ukraineUnnamed: 5', 'losses_ukraineabandoned',
    'losses_ukraineabandoned and destroyed', 'losses_ukrainecaptured', 'losses_ukrainecaptured and destroyed',
    'losses_ukrainedamaged', 'losses_ukrainedamaged and abandoned', 'losses_ukrainedamaged and captured',
    'losses_ukrainedamaged beyond economical repair', 'losses_ukrainedestroyed',
    'losses_ukrainescuttled to prevent capture by Russia', 'losses_ukrainesunk',
    'losses_ukrainesunk but raised by Russia'
]

# Compute the correlation matrix
corr_matrix = RUS_UKR[RUS_UKR_Columns].corr()

# Iterate over the correlation matrix and print the correlated categories
for col1 in corr_matrix.columns:
    for col2 in corr_matrix.columns:
        if col1 != col2:
            correlation = corr_matrix.loc[col1, col2]
            if correlation >= 0.7 or correlation <= -0.7:
                print(f"Correlation between {col1} and {col2}: {correlation}")
```
